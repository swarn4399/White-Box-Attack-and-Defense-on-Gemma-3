```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch, copy, evaluate
from datasets import load_dataset

# ── Load base model once ──────────────────────────────────────────────────────
model = AutoModelForCausalLM.from_pretrained(model_path, dtype=torch.bfloat16).to("cuda").eval()
model.config.use_cache = False
tokenizer = AutoTokenizer.from_pretrained(model_path)

# ── Config ────────────────────────────────────────────────────────────────────
LAYER_INDICES = [0, 8, 17]          # first, middle, last
METHODS       = ["zeroed", "random"] # add "retrained" later once you have checkpoints
PROMPTS       = ["Tell a story:", "Factor x^2+3x+2:"]

CKPT_PATHS = {                       # fill these in once you have checkpoints
    0:  "checkpoints/layer_0.pt",
    8:  "checkpoints/layer_8.pt",
}

# ── Dataset for perplexity ────────────────────────────────────────────────────
ds    = load_dataset("wikitext", "wikitext-2-raw-v1", split="test")
texts = ["\n".join(ds["text"])]
ppl_metric = evaluate.load("perplexity", module_type="measurement")

# ── Helper functions ──────────────────────────────────────────────────────────
def apply_method(new_layer, method, layer_idx):
    if method == "zeroed":
        excluded_params = set()
        for name, param in new_layer.named_parameters():
            if "layernorm" in name.lower() or "layer_norm" in name.lower():
                excluded_params.add(param.data_ptr())
        with torch.no_grad():
            for param in new_layer.parameters():
                if param.data_ptr() not in excluded_params:
                    param.data.zero_()

    elif method == "random":
        def reset_weights(m):
            if hasattr(m, 'reset_parameters'):
                m.reset_parameters()
        with torch.no_grad():
            new_layer.apply(reset_weights)

    elif method == "retrained":
        ckpt_path = CKPT_PATHS[layer_idx]
        with torch.no_grad():
            state_dict = torch.load(ckpt_path, map_location="cuda")["model_state_dict"]
            new_layer.load_state_dict(state_dict, strict=True)

    return new_layer

def get_ppl(model):
    save_dir = "model_temp_save"
    model.save_pretrained(save_dir)
    tokenizer.save_pretrained(save_dir)
    with torch.no_grad():
        res = ppl_metric.compute(
            model_id=save_dir, data=texts,
            add_start_token=True, batch_size=4, max_length=512
        )
    return res["mean_perplexity"]

def get_generation(model, prompt):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
    gen_ids = model.generate(
        input_ids, max_new_tokens=50,
        do_sample=True, temperature=1.0, use_cache=True
    )
    return tokenizer.decode(gen_ids[0], skip_special_tokens=True)

# ── Baseline first ────────────────────────────────────────────────────────────
print("=" * 60)
print("BASELINE (no modification)")
print("=" * 60)
baseline_ppl = get_ppl(model)
print(f"PPL: {baseline_ppl:.4f}")
for prompt in PROMPTS:
    print(f"\nPrompt: {prompt}")
    print(get_generation(model, prompt))

results = {"baseline": {"ppl": baseline_ppl}}

# ── Main loop ─────────────────────────────────────────────────────────────────
for layer_idx in LAYER_INDICES:
    for method in METHODS:
        print("\n" + "=" * 60)
        print(f"Layer {layer_idx} | Method: {method}")
        print("=" * 60)

        # Reload fresh base model for each run so modifications don't bleed over
        test_model = AutoModelForCausalLM.from_pretrained(
            model_path, torch_dtype=torch.bfloat16
        ).to("cuda").eval()
        test_model.config.use_cache = False

        new_layer = copy.deepcopy(test_model.model.layers[layer_idx]).to("cuda").eval()
        new_layer = apply_method(new_layer, method, layer_idx)

        with torch.no_grad():
            test_model.model.layers[layer_idx] = new_layer

        # Generations
        for prompt in PROMPTS:
            print(f"\nPrompt: {prompt}")
            print(get_generation(test_model, prompt))

        # Perplexity
        ppl_val = get_ppl(test_model)
        print(f"\nPPL: {ppl_val:.4f}")
        results[f"layer{layer_idx}_{method}"] = {"ppl": ppl_val}

        # Free VRAM before next iteration
        del test_model, new_layer
        torch.cuda.empty_cache()

# ── Summary table ─────────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("SUMMARY")
print("=" * 60)
print(f"{'Config':<25} {'PPL':>12}")
print("-" * 40)
print(f"{'baseline':<25} {results['baseline']['ppl']:>12.4f}")
for layer_idx in LAYER_INDICES:
    for method in METHODS:
        key = f"layer{layer_idx}_{method}"
        print(f"{'layer '+str(layer_idx)+' '+method:<25} {results[key]['ppl']:>12.4f}")
```
